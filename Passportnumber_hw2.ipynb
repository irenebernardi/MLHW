```python
df = pd.read_csv('/Users/irenebernardi/Desktop/ML /homeworks/hw2/ex2data1.csv')

#one hot encoding 
edf= pd.get_dummies(df)

# Splitting the data into training and testing sets
x = edf.iloc[:, :-1]  #features 
y = edf.iloc[:, -1] #label 
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.2)



#double check that had to one hot encode y label as well???
```

```python
#information gain ( = entropy before split  entropy after split)
tree = dtc(criterion = "entropy", max_depth = 2)
#TODO: do this on paper and submit this part as well 
tree.fit(x_train, y_train)
#TODO: CHANGE CLASSIFIER VLAUES W ONRE HOT ENCODED VALUES 
#plotting the tree

plt.figure(figsize=(12, 8))
plot_tree(tree, filled=True, class_names=['projection neurons', 'interneurons'], rounded=True)
plt.title("Decision Tree")
plt.show()
```
